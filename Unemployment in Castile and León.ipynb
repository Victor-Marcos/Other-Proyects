```python
labels = []
j = 0
for i in Palencia_data["FECHA"]:
    if j % 20 == 0:
        labels.append(i)
    j = j + 1
plt.plot(Palencia_data['FECHA'], Palencia_data['TOTAL'], label = "Total")
plt.plot(Palencia_data['FECHA'], Palencia_data['VARON'], label = "Men")
plt.plot(Palencia_data['FECHA'], Palencia_data['MUJER'], label = "Women")
plt.xlabel('Date')
plt.ylabel('Total unemployed')
ticks_locations = range(0, len(Palencia_data["FECHA"]), 20)
plt.xticks(ticks_locations, labels, rotation=45)
plt.legend()
```
